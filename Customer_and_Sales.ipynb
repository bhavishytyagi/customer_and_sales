Number of Customers by their payment method

```python
pc=new.groupby(['payment_method'],as_index=False)['customer_id'].count()
pc
```

```python
plt.pie(data=pc,x='customer_id',labels='payment_method',autopct='%1.1f%%',explode=(0.02,0.02,0.02))
plt.gcf().gca().add_artist(plt.Circle((0,0),0.0,fc='white'))
plt.title('Payment Category')
plt.show()
```

Number of sales by Gender

```python
gen_sale=new.groupby(['gender'],as_index=False)['customer_id'].count()
gen_sale
```

```python
ax=sns.barplot(x='gender',y='customer_id',data=gen_sale)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
gen_amt=new.groupby(['gender'],as_index=False)['price'].sum()
gen_amt
```

```python
ax=sns.barplot(y='gender',x='price',data=gen_amt,palette=['Red','Orange'])
for bars in ax.containers:
  ax.bar_label(bars)
```

age_bucket

```python
def conditions(i):
    if i <=20: return '0-20'
    if i > 20 and i <=30: return '20-30'
    if i > 30 and i <=40: return '30-40'
    if i > 40 and i <=50: return '40-50'
    if i > 50 and i <=60: return '50-60'
    if i > 60 and i <=70: return '60-70'
    if i > 70: return '>70'
    else : return 'Unavailable'

new['age_gp']=new['age'].apply(conditions)
```

```python
new
```

Number of sales according to age

```python
age=new.groupby(['age_gp'],as_index=False)['customer_id'].count()
age
```

```python
ax=sns.barplot(x='age_gp',y='customer_id',data=age)
for bars in ax.containers:
  ax.bar_label(bars)
```

Sales acc to categories

```python
cat_sales=new.groupby(['category'],as_index=False)['price'].sum()
cat_sales
```

```python
plt.figure(figsize=(12,4))
ax=sns.barplot(x='category',y='price',data=cat_sales)
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

Sales acc. to date

```python
new['year']=pd.DatetimeIndex(new['invoice_date']).year
new
```

```python
sales_date=new.groupby(['year'],as_index=False)['customer_id'].count()
sales_date
```

```python
plt.pie(data=sales_date,x='customer_id',labels='year',autopct='%1.1f%%',pctdistance=0.75,explode=(0.05,0.05,0.05))
cc=plt.Circle((0,0),0.50,fc='white')
plt.gcf().gca().add_artist(cc)
plt.title('Yearwise Sales')
plt.show()
```

no. of sales per shopping mall

```python
mall=new.groupby(['shopping_mall'],as_index=False)['customer_id'].count().sort_values(by='customer_id',ascending=False)
mall
```

```python
plt.figure(figsize=(14,4))
ax=sns.barplot(y='shopping_mall',x='customer_id',data=mall,palette=['magenta'])
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```
